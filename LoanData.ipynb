## categoric to numeric
To deal with the rest missing values we have to convert categorical columns to the numeric. We do it by dummies method as well as by defining some functions.
Let's make it easy to deal with loan_status create a new column where 1 will be for Fully paid loans and 0 otherwise. As there are no explanation whit "does not meet the credit policy...", I found out that it means something like the credit policy was updated after this loan was listed but before it issued. So, we will base on status given here and therefore, define

```python
def converter(loan_status):
        if (loan_status=='Fully Paid') | (loan_status =='Does not meet the credit policy. Status:Fully Paid'):
            return 1
        else:
            return 0
```

```python
loans['loan_status'] = loans['loan_status'].apply(converter)
```

The same we will do with other columns, and in int_rate column we just delete the % sign

```python
loans["int_rate"] = loans["int_rate"].str.rstrip("%").astype("float")
```

```python
loans["revol_util"] = loans["revol_util"].str.rstrip("%").astype("float")
```

```python
dummy_loans = pd.get_dummies(loans['term'],drop_first=True) # to delete 36 months column 
loans = pd.concat([loans, dummy_loans], axis=1)
loans = loans.drop([ "term"], axis=1)
```

### Business case:


to show how the model works we print out results of our prediction according to the interest rate. As the higher interest rate is probably more attractive for investors, but usually they are more risky, i.e. the riskier the borrower, the higher the interest rate. With interest rate in mind, we can then determine if the borrower is eligible for the loan.
Moreover, in our classification problem False Negatives are a lot more expensive than False Positives (to give the loan with probability not to get payment back for investors are more expansive than to miss the opportunity to give the loan potentially trusty borrower). So in our case, high precision (TP/(TP+FP)) is more important than high recall (TP/(TP+FN)).

## SMOTE (Synthetic Minority Oversampling Technique)

Do not oversample before splitting into training and test no to bleed information from the test set into the training. However, because the SMOTE algorithm uses the nearest neighbors of observations to create synthetic data, it still bleeds information. If the nearest neighbors of minority class observations in the training set end up in the test set, their information is partially captured by the synthetic data in the training set. Since we split the data randomly, we expect this happen. As a result, the model will be better able to predict validation set values than completely new data.
