```python
import pandas as pd
```

```python
data = pd.read_csv('loans.csv')
```

```python
data.head()
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data.drop('loan_status',axis=1))
```

```python
scaled_features = scaler.fit_transform(data.drop('loan_status',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=data.columns[:-1])
df_feat.head()
```

```python
X = df_feat
```

```python
y = data['loan_status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import tensorflow as tf
```

```python
df_feat.columns
```

```python
feat_col = []
for col in df_feat.columns:
    feat_col.append(tf.feature_column.numeric_column(col))
```

```python
feat_col
```

```python
classifier = tf.estimator.DNNClassifier(hidden_units=[10, 20, 10], n_classes=2,feature_columns=feat_col)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,shuffle=True)
```

```python
classifier.train(input_fn=input_func,steps=100)
```

```python

pred_fn = tf.estimator.inputs.pandas_input_fn(x = X_test, batch_size=len(X_test), shuffle=False)
```

```python
note_predictions = list(classifier.predict(input_fn=pred_fn))
```

```python
note_predictions = list(classifier.predict(input_fn=pred_fn))
```

```python
note_predictions[0]
```

```python

final_pred = []
for pred in note_predictions:
    final_pred.append(pred['class_ids'][0])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, final_pred))
```

```python
print(confusion_matrix(y_test, final_pred))
```

```python
confusion_loans = pd.DataFrame(confusion_matrix(y_test,final_pred),
             columns=["Predicted Charged off", 'Predicted Fully paid'],
             index = ["Charged off", 'Fully paid'])

print(confusion_loans)
```

```python
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(y_test, final_pred, normalize=True)
```

```python
loan_pred_dnn = pd.DataFrame({'Actual': y_test, 'Predicted': final_pred})
```

```python
loan_pred_dnn.reset_index(inplace=True)
```

```python
inds = np.random.permutation(np.arange(len(X_test)))
prognose_table_dnn = sorted( zip(data.loc[inds, 'int_rate'],loan_pred_dnn.loc[inds,'Actual'] == loan_pred_dnn.loc[inds,'Predicted']),key=lambda x: (-x[0], -x[1]))
```

```python
line = 0
for l in prognose_table_dnn:
    print('Interest_rate = {0}%, model_correct = {1}'.format(*l))
    line += 1
```
